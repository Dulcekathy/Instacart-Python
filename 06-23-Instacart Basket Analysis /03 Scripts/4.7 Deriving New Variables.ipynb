## Using If-Statements with User-Defined Functions

```python
# subset of the dataframe for the first one million rows.
df = orders_products_merged[:1000000]
```

```python
df.shape
```

```python
#Define price label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply price label on dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check values in your new column
df['price_range'].value_counts(dropna = False)

#Notice that you don’t have any rows with the “High-range product” label. 
#This means there aren’t any products within the dataframe greater than $15.
```

# Step 4 Create busiest period of the day column

```python
# Check frequency of values in the "order_hour_of_day" column 
orders_products_merged['order_hour_of_day'].value_counts()
```

### The client wants orders grouped into three different periods of time. I can divide the 24 hrs into three periods of time (Most orders = 10, 11, 14, 15, 13, 12, 16, 9),  (Average orders = 17, 8, 18, 19, 20, 7, 21, 22), and (Fewest orders = 23, 6, 0, 1, 5, 2, 4, 3).

```python
# Creating a for-loop if statement using the breakdown above.
hour_of_day = []

for value in orders_products_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_day.append("Fewest orders")
  else:
    hour_of_day.append("Average orders")
```

```python
# Adding result to dataframe as "busiest_period_of_day" column
orders_products_merged['busiest_period_of_day'] = hour_of_day
```
